```python
import numpy as np
#combines values from heart disease history and other major disease history to one colum as patient with major disease history
#setting values for people with heart disease and with any other major disease as 3 and with heart disease as 2 and other major disease as 1 and no major disease or heart disease as 0
disease_history = np.where(
    (insure_data['patient_has_heart_disease_history'] == 1) & (insure_data['patient_has_other_major_disease_history'] == 1),
    3,
    np.where(
        (insure_data['patient_has_heart_disease_history'] == 1) & (insure_data['patient_has_other_major_disease_history'] ==0 ),
        2,
        np.where(
            (insure_data['patient_has_heart_disease_history'] == 0) & (insure_data['patient_has_other_major_disease_history'] == 1),
            1,
            0
        )
    )
)
print(disease_history)
```
